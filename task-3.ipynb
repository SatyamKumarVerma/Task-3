```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\asus\\Desktop\\github\\task\\householdtask3.csv')
```

```python
df.head(10)
```

```python
#Scatter plot with year against own
plt.scatter(df['year'], df['own'])

#Adding title to the plot
plt.title("SCATTER PLOT")

#Setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Line Chart with year against own
plt.plot(df['year'])
plt.plot(df['own'])

#Adding title to the plot
plt.title("Line Chart")

#Setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Bar Chart with year against own
plt.bar(df['year'], df['own'])

#Adding title to the plot
plt.title("Bar Chart")

#Setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Histogram
plt.hist(df['income'])

plt.title("HISTOGRAM")

plt.show()
```
